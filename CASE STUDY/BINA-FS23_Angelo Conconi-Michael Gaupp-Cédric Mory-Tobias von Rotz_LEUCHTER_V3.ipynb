# Management Summary
---

**Situation & Entscheidungsbedarf**
<p align="justify"> Die Leuchter Software Engineering AG ist ein Softwareentwicklungsunternehmen, welches Individualsoftware sowie eigene Softwareprodukte im Bereich der automatisierten Dokumentenerstellung entwickelt. Hierbei verwendet die Leuchter Software Engineering AG für die Auftragsplanung und das Projektcontrolling das Issue-Tracking-Tool (vgl. Ticketing-System) Gemini vom Software-Hersteller Countersoft. Mittels Gemini werden die Aufwandschätzungen und Aufwandsrapportierungen für alle Projekte vorgenommen.</p>
<br>
<p align="justify"> Im Rahmen einer internen Analyse zur Business-Excellence wurden Optimierungspotenziale im Projektcontrolling identifiziert. Insbesondere wurde der Bedarf an Indikatoren geäussert, welche Aussagen zu Risiken in Form von Budgetüberschreitungen und der Rentabilität von Projekten als auch Einschätzungen zur Performance von Entwicklerressouren ermöglichen. Hierfür bietet sich die Nutzung von Gemini an, da dieses in der zugehörigen Datenbank bereits, alle Daten die von Interesse sind, führt. Es gilt diesbezüglich das Gemini auf die Eignung, den Erfüllungsgrad der Nutzenpotenziale hin und die Verwendung für den Projektcontrolling-Zweck zu überprüfen.</p>
<br>

**Vorgehen & Handlungsoptionen**
<p align="justify"> Für die vorliegende, offene Zielsetzung bietet es sich an, ein etabliertes Data-driven Decision making (DDDM) Framework zu verwenden, um den Informationsgewinn strukturiert-inkremetell, ergo teleologisch zu verfolgen. Hinsichtlich der gegebenen Situation wurde diesbezüglich auf das 5-Schritte-Modell "From Data to Decisions" von Chartered Professional Accountants Canada (CPA) zurückgegriffen, um die Eignung von Gemini zu evaluieren. Hierfür wurde vom Studierendenteam ein Setup im Sinne eines Testings nach dem Modell von CPA bereitgestellt. Gemäss Aufgabenstellung wurden lediglich die Schritte 1-4 erarbeitet. 

<br><p align="justify"> In einem ersten Schritt wurden die Bedürfnisse auf Basis der Optimierungspotenziale überprüft und den zweckgemässen Informationsbedarf abgeleitet. Hierfür wurde ein Experteninterview mit einem Vertreter des Management Boards der Leuchter Software Engineering AG abgehalten. Anschliessend wurde eine objektive Zielsetzung erarbeitet, welche sinngemäss als Bewertungskatalog zur Eignungsprüfung des Geminis dient (vgl. Kapitel Step 1). Das Studierendenteam wurde im Auftrag der Leuchter Software Engineering AG mit der Aufgabe betraut, die Eignungsprüfung als vorliegende Abhandlung zum Gemini vorzunehmen.

<br><p align="justify"> In einem zweiten Schritt wurde die, gemäss erstem Schritt, relevante Datengrundlage beschafft. Hierbei galt ein besonderer Fokus der Datenmanipulation und späteren -analyse der Datensets. Nach anfänglichen Erschwernissen setzte das Studierendenteam einen SQL-Server in Azure mit dem Back-Up der produktiven Gemini-Datenbank auf, welche als Datenquelle dient (vgl. Kapitel Step 2). Abklärungen zu weiteren Datenquellen wurden unternommen, als weitere Datenquelle kam das Enterprise-Ressourcing-Planning-Systems der Leuchter Software Engineering in Frage, welches allerdings infolge von Vertraulichkeitsthemen nicht verwendet werden konnte.

<br><p align="justify"> In einem dritten Schritt wurden die gegebenen Daten hinsichtlich  Relevanz und Eignung zur Zielerfüllung überprüft. Im Weiteren stand die spätere Belastbarkeit der Daten im Fokus. Diesbezüglich wurden umfangreiche Analysen vorgenommen, um von gegebenen Daten, auf den zweckgerichteten Informationsbedarf der Leuchter Software Engineering AG zu schliessen. Weitere Kalkulationen wurden diesbezüglich notwendig (vgl. Kapitel Step 3).</p>

<br>
<p align="justify"> In einem vierten Schritt wurden die analysierten, konkretisierten und aufbereiteten Daten visuell dargestellt. Dabei wurde in Rücksprache mit der Leuchter Software Engineering ein Cockpit in Form eines Microsoft PowerBI Dashboardd erstellt, welches ebenfalls auf demselben Applikationsserver (vgl. SQL-Server) installiert wurde (vgl. Kapitel Step 4).</p>
<br>

**Ergebnis & Empfehlung**
<p align="justify"> Der Informationsbedarf der Leuchter Software Engineering AG konnte vollumfänglich auf Basis der Gemini-Datenbank abgebildet und in Form eines funktionsfähigen PowerBI Cockpits visualisiert werden. Die Leuchter Software Engineering AG wird vom Studierendenteam angehalten, die Lieferergebnisse auf ihre Bedürfnisse hin zu überprüfen. Anpassungen, Implementation und Integration der Ergebnisse, welche aus dem Testing zum Live-Betrieb (Produktivbetrieb) führen, obliegen der Leuchter Software Engineering AG. Das Studierendenteam empfiehlt die weitere Verfolgung des Projektvorhabens auf Basis der positiven Lieferobjekte auszubauen.</p>
<br>


**Nächste Schritte**
<p align="justify"> Im Anschluss an die Re-Evaluation der Liefergebenisse wird der Leuchter Software Engineering AG empfohlen, den Mitteleinsatz im Live-Betrieb zu begleiten und kontinuierlich auf die langfristige Zielerreichung hin zu überprüfen.</p>

# **Einleitung**
---

**Ausgangslage**

<p align="justify"> Die Leuchter Software Engineering AG (SE) wurde 2012 gegründet und ist in der Stadt Luzern ansässig. Das Unternehmen ist im Bereich der Softwareentwicklung tätig und bedient ihre Kunden mit individuellen, massgeschneiderten Softwarelösungen als auch mit ihren Software Produkten wie "Docugate" und "Signgate" (Leuchter AG, 2023).

<p align="justify"> Für die Administration und das Controlling aller Aufträge nutzt die Leuchter Software Engineering AG ein Issue-Tracking-Tool namens Gemini des Software-Herstellers Countersoft. In Gemini werden insbesondere die Aufwandschätzung und Aufwandsrapportierung innerhalb von Issues für: Neuentwicklungen, Software-Releases, Nachträgen und Garantiearbeiten erfasst und darin abgebildet. 

<br>**Anstoss zur Untersuchung**

<p align="justify"> Im Rahmen einer internen Analyse in puncto Business-Excellence wurden Optimierungspotenziale, insbesondere im Projektcontrolling, identifiziert. Hierbei wären Indikatoren wünschenswert, welche die Rentabilität von Projekten aufzeigt und die Einschätzungen zu der Performance von Entwicklerteams ermöglicht. 
<br>
<br>Zwar werden im Gemini breite Geschäftsdaten genutzt; diesbezüglich stellt Gemini auch umfangreiche Funktionalitäten für die Geschäftstätigkeit bereit. Diese Daten werden aktuell aber vorwiegend für die administrativen Belange im operativen Projektmanagement genutzt. Entsprechend sollen die Daten aus Gemini (ggfs. in Kombination mit dem weiteren Tool-Set der SE) zur Nutzung im Projektcontrolling überprüft werden.

<br>**Grundlegende Zielsetzung**

<p align="justify"> Die SE beabsichtigt übergeordnet, Informationen zu den Arbeitspaketen (Wording Gemini: Issues, sinngem. Aufträge) weiterführend zu analysieren, um frühzeitige Projektrisiken in Hinsicht auf Budget- oder Terminüberschreitungen erkennen zu können. Hierzu sollen, sofern möglich, Auswertungen auf Projekt- und Mitarbeiter-Ebene visualisiert werden. Optional soll überprüft werden, ob Machine Learning für das Reporting verwendet werden kann und daraus resultierende Forecasts auf Projektebene, die Arbeit der Projektleiter erleichtern kann. Die Forecasts sollen zu einer qualitativ-besseren Planung von Projekten beitragen und so die Ressourcenplanung der Mitarbeitenden auf Projekte optimieren.

<br><p align="justify"> Die Zielsetzung bedarf einer weiterführenden Überprüfung, um teleologische Lieferobjekte bzw. Artefakte bereitstellen zu können.

<br>**Methodik**

<p align="justify"> Auf Basis der grundlegenden Zielsetzung bietet es sich an, ein am Markt etabliertes Framework für data-driven-decision-making (DDDM) zu nutzen. Hierfür wird auf das Framework "From Data to Desicions" von Chartered Professional Accountants Canada (CPA) zurückgegriffen (Marr, 2020). Das Framework eignet sich ideal für die Ausgangslage des Vorhabens und führt in fünf Schritten an die datengetriebene Entscheidungsfindung, wie sie für die SE nützlich wäre, heran.

*Für das Projektvorhaben werden aus zeitlichen Einschränkungen lediglich die Schritte 1 - 4 durchgeführt.*


<br>**Alternativen**
<p align="justify"> DDDM ist ein aktuelles, mit grossem Interesse, verfolgtes Forschungsgebiet. Nebst dem Framework von CPA stehen weitere, ähnliche Frameworks wie z. B. der DDDM-Guide von tableau bereit. (tableau, 2023). Diese sind sich in den Ansätzen sehr ähnlich und unterscheiden sich vorwiegend in der Anzahl Schritte. Da kein signifikanter Mehrwert durch dieses Framework besteht, wird diesbezüglich das DDDM-Framework von CPA weiterverfolgt.

<br> **Vorgehen**

Für das Vorhaben wurde das induktive Vorgehen gemäss Framework gewählt. Dieses verfolgt die folgenden Schritte. 

<br>**Step 1 - "Defining objectives and information needs"**
<p align="justify"> Im ersten Schritt erfolgt eine differenzierte Betrachtung der grundlegenden Zielsetzung. Hierbei liegt der Fokus auf der präzisen Definition von Zielen und den erforderlichen Informationen. Dieses Vorgehen gewährleistet, dass die richtigen Daten erhoben werden und daraus relevante Informationen abgeleitet werden können. Dadurch wird eine adäquate Erfüllung der Anforderungen der SE angestrebt.

<br>**Step 2 - "Collecting data"**
<p align="justify"> Im nächsten Schritt erfolgt die Datensammlung. Dabei ist es von grosser Bedeutung, relevante und valide Daten zu berücksichtigen. Des Weiteren werden die quantitativen Aspekte der Daten, wie beispielsweise ihre Verfügbarkeit und Menge, sorgfältig beachtet.

<br>**Step 3 - "Analyzing data"**
<p align="justify"> Im dritten Schritt erfolgt die Datenanalyse, welche darauf abzielt, die Zuverlässigkeit und den damit verbundenen Erkenntnisgewinn zu gewährleisten. Da die vorliegenden Daten Projekte oder Issues betreffen, eignen sich zur Anwendung eher quantitative Analysemethoden wie Data Mining oder Regression. Die konkrete Methode wird nach der Datensammlung festgelegt.

<br>**Step 4 - "Presenting information"**
<p align="justify"> Im vierten Schritt werden die aus den Daten abgeleiteten Informationen präsentiert, um den spezifischen Nutzungsanforderungen der SE gerecht zu werden.

Aufgrund dieser Tatsache, ist eine Weiterverfolgung von Z4 aktuell nicht möglich, da der Terminplan nicht für die Auswertung verwendet werden kann. 

<p align="justify"> Desweiteren konnten keine ergänzenden sinnvollen externen Quellen für die vorliegende Datenbasis ermittelt und herbeigezogen werden.

<br> **Datenprüfung**</br>
*(→"Considering data ownership"; →"Considering data privacy"; →"Balancing structured data with unstructured data")*
<p align="justify"> Als Nächstes wurde eine Klärung zu den Daten hinsichtlich Eigentum und Privatsphäre vorgenommen. 

<p align="justify"> Die Daten sind im Eigentum der SE mit den entsprechenden Eigentums- und Besitz-Rechten behaftet. Nach Rücksprache mit der SE handelt es sich bei beiden Datensätzen um sensible Daten. Einblicke über Umsatzzahlen, Gewinn) nicht ausgehändigt noch offengelegt werden. Die Sekundärquelle konnte daher im weiteren Vorhaben nicht berücksichtigt werden. Folglich bezieht sich die vorliegende Abhandlung lediglich auf die primäre Datenquelle SE_Gemini_Prod ohne ergänzende, externe Datensätze.

Unstrukturierte Daten (im Sinne von Metainformationen) stehen aus dem Interview bereit, beziehen sich aber nicht effektiv auf den Datensatz und können resultativ schlussgefolgert werden. Daher scheint eine Nivellierung durch eine fehlende Bezugsbasis, zwischen unstrukturierten und strukturierten Daten für die vorliegende Abhandlung, als nicht gegeben.

```python
# Code-Snippet 400
# Sprache: Structured Query Language
# -----------------------------------------------------------------------------------------------------

# 1. Erstellung der VIEW 
USE [SE_Gemini_PROD]
GO

/****** Object: View [dbo].[SE_PowerBI] Script Date: 14.05.2023 14:28:25 ******/
SET ANSI_NULLS ON
GO

SET QUOTED_IDENTIFIER ON
GO

ALTER view [dbo].[SE_PowerBI]
AS

SELECT dbo.gemini_issues.fixedinversionid
, dbo.gemini_versions.versionnumber
, dbo.gemini_versions.versionname
, dbo.gemini_projects.projectcode
, dbo.gemini_projects.projectname
, dbo.gemini_issuestatus.statusdesc
, dbo.gemini_issues.issueid
, dbo.gemini_issues.estimatehours
, dbo.gemini_issues.estimateminutes
, dbo.gemini_issues.loggedhours
, dbo.gemini_issues.loggedminutes
, dbo.gemini_users.username
, dbo.gemini_users.firstname
, dbo.gemini_users.surname
, dbo.gemini_users.emailaddress
, dbo.gemini_users.roles
, dbo.gemini_components.componentname
, dbo.gemini_components.componentdesc
, dbo.gemini_issues.created	
, dbo.gemini_issues.closeddate	

FROM dbo.gemini_issues INNER JOIN
 dbo.gemini_issuestatus ON dbo.gemini_issues.issuestatusid = dbo.gemini_issuestatus.statusid INNER JOIN
 dbo.gemini_projects ON dbo.gemini_issues.projectid = dbo.gemini_projects.projectid INNER JOIN
 dbo.gemini_issuecomponents ON dbo.gemini_issues.issueid = dbo.gemini_issuecomponents.issueid INNER JOIN
 dbo.gemini_versions ON dbo.gemini_issues.fixedinversionid = dbo.gemini_versions.versionid INNER JOIN
 dbo.gemini_issueresources ON dbo.gemini_issues.issueid = dbo.gemini_issueresources.issueid INNER JOIN
 dbo.gemini_users ON dbo.gemini_issueresources.userid = dbo.gemini_users.userid INNER JOIN
 dbo.gemini_components ON dbo.gemini_issuecomponents.componentid = dbo.gemini_components.componentid
GO
```

<u>Zielbetrachtung Z3 & Z3.1</u>

In einem nächsten Schritt wurden die gekoppelten Ziele betrachtet:

<li> Z3.1 - Budgetüberschreitung 
</li></li>

<p align="justify"> Die in der View geladenen Tabellen, verfügen über die geschätzten Stunden ("estimatedhours" sowie "estimatedminutes") in zwei separaten Spalten. Zwecks einfacher Vergleichbarkeit der Projekte sollten die Informationen in eine Variable überführt werden, gemäss Indikator (vgl. Tabelle 1) als Stunden.

<p align="justify"> Folglich galt es diese zu addieren und als neue Spalte abzubilden "calcEstimated". Dabei galt es zu beachten, dass die Minuten unweigerlich zu Dezimalzahlen führen, ergo musste der Datentyp FLOAT gesetzt werden, daher der Multiplikator *1.0 (vgl. Code-Snippet 410).

```python
# Code-Snippet 410
# Sprache: Structured Query Language
# -----------------------------------------------------------------------------------------------------

# Bildung Ersatzvariable geschätzter Stunden
, (dbo.gemini_issues.estimatehours * 1.0) + (dbo.gemini_issues.estimateminutes / 60.0) AS 'calcEstimated'
```

<p align="justify"> Im nächsten Schritt galt es dasselbe Prinzip für die gebuchten Stunden ("loggedhours" sowie "loggedminutes") vorzunehmen. Bereitgestellt wurde die Summen-Variable "calcLogged" (vgl. Code-Snippet 420).

```python
# Code-Snippet 420
# Sprache: Structured Query Language
# -----------------------------------------------------------------------------------------------------

# Bildung Ersatzvariable gebuchte Stunden
, (dbo.gemini_issues.loggedhours * 1.0)+(dbo.gemini_issues.loggedminutes / 60.0) AS 'calcLogged'
```

<p align="justify"> In Bezug zur Zielbetrachtung Z1 & Z3 galt es nun die Budgetüberschreitung ausweisen zu können. Der Berechnung liegt somit eine Differenz zu Grunde, welche nach der Logik von Reststunden bzw. verbleibenden Stunden ausgewiesen wird. Entsprechend wurde der Query eine neue Spalte angelegt, welche die Differenz von SOLL(geschätzten Stunden) zu IST (gebuchten Stunden) aufzeigt. Ist der Wert positiv oder 0, so verlief das Issue rentabel. Die Differenz wurde als Spalte "calcDiffHours festgehalten (vgl. Code-Snippet 430).



```python
# Code-Snippet 430
# Sprache: Structured Query Language
# -----------------------------------------------------------------------------------------------------

# Bildung Differenz SOLL-IST-Stunden
, ((dbo.gemini_issues.estimatehours,0.0) * 1.0) + ((dbo.gemini_issues.estimateminutes,0.0) / 60.0) - ((dbo.gemini_issues.loggedhours,0.0)* 1.0)+((dbo.gemini_issues.loggedminutes,0.0) / 60.0) AS 'calcDiffHours'
```

<p align="justify"> Nach einer schnellen Prüfung des Datensatz auf Basis der Kalkulationen von Code-Snippets 410 - 430 wurde ein Problem identifiziert. Es existieren Projekte bzw. Issues, welche zwar geschätzt, aber vom Kunden nie bestellt wurden. Ebenfalls möglich wären auch Issues, welche nicht sauber gepflegt wurden und entsprechend keine Zeiteingaben beinhalten. So wurde der Sachverhalt über die Abfrage nach "<> 0" verdeutlicht (vgl. Code-Snippet 440).


```python
# Code-Snippet 440
# Sprache: Structured Query Language
# -----------------------------------------------------------------------------------------------------

# Prüfung auf Null-Eintragungen
SELECT * FROM dbo.SE_PowerBI WHERE calcDiffHours <> 0.0
```

<p align="justify"> Diesbezüglich mussten die Kalkulationen diesen Faktor berücksichtigen. Hierfür bot es sich an, die Kalkulationen mit der SQL-Funktion "COALESCE" berechnen zu lassen. Die Skalar-Funktion "COALESCE" ermöglicht, fehlende Eintragungen als Nullwerte (engl. "null", "nullable") auszuweisen und diese nicht in die Kalkulation mit dem Wert "0" zu berücksichtigen (engl. "not null"). Diese Funktion führt eine Datenbereinigung durch und behebt die Problematik (vgl. Microsoft, 2023a).

<p align="justify"> Die ursprünglichen Queries wurden in untenstehendem Code-Snippet mit der COALESCE-Funktion ergänzt (vgl. Code-Snippet 450).

```python
# Code-Snippet 450
# Sprache: Structured Query Language
# -----------------------------------------------------------------------------------------------------

# für calcEstimated
, (COALESCE(dbo.gemini_issues.estimatehours,0.0) * 1.0)+(COALESCE(dbo.gemini_issues.estimateminutes,0.0)/60.0) AS 'calcEstimated'

# für calcLogged
, (COALESCE(dbo.gemini_issues.loggedhours,0.0)* 1.0)+(COALESCE(dbo.gemini_issues.loggedminutes,0.0)/60.0) AS 'calcLogged'

# für calcDiffHours
, (COALESCE(dbo.gemini_issues.estimatehours,0.0) * 1.0)+(COALESCE(dbo.gemini_issues.estimateminutes,0.0) / 60.0) - (COALESCE(dbo.gemini_issues.loggedhours,0.0) * 1.0)+(COALESCE(dbo.gemini_issues.loggedminutes,0.0) / 60.0) AS 'calcDiffHours'
```

Die betroffenen Visualisierungen konnten mithilfe von Rechtecken abgedeckt werden und deren Transparenzeinstellungen über ein zzgl. Measure im PowerBI. In Ergänzung dazu wurde ein String eingeblendet, welcher den Nutzer des PowerBI Cockpits auffordert, Filtereinstellungen vorzunehmen. Die Details sind den untenstehenden Code-Snippets (vgl. Code-Snippet 610-612) und dem Screenshot (vgl. Abbildung 14 & 15 zu entnehmen).

```python
# Code-Snippet 610
# Sprache: Microsoft Power BI / Data Analysis Expressions (DAX)
# Überprüfung ob Filter gesetzt sind (Variable Check Filtered)
# -----------------------------------------------------------------------------------------------------

# neues Measure "Check Filtered", überprüft ob Filter in Datentabelle gesetzt
Check Filtered = ISFILTERED(SE_PowerBI) -- returns True if Filters set in Table "SE_PowerBI"
```

```python
# Code-Snippet 611
# Sprache: Microsoft Power BI / Data Analysis Expressions (DAX)
# Transparenzsteuerung der Überdeckung (Variable Make Transparent, letztes Crum "00")
# -----------------------------------------------------------------------------------------------------

# neues Measure setzt Füllung anhand Check Filtered fest, transparent wenn gesetzt
Make Transparent = 
      IF(
         [Check Filtered],
         "#FFFFFF00", -- returns transparent if Check Filtered is True
         "White"
      )
```

```python
# Code-Snippet 612
# Sprache: Microsoft Power BI / Data Analysis Expressions (DAX)
# Notification wenn keine Filter gewählt sind (Variable Message)
# -----------------------------------------------------------------------------------------------------

# Gibt String zurück wenn keine Filter gesetzt sind
Message = 
   IF(
      [Check Filtered], -- Check Filtered returns True or False based on filter settings
      "", -- null/nothing when Check Filtered returns True, Else write text
      "Bitte Filter rechts setzen."
   )
```

<u>Wertebereich Performance</u>
<br><p align="justify"><i>Problemstellung:</i> Der Wertebereich des Tachometer (vgl. Visualisierungen Performance) werden durch PowerBI automatisch berechnet. Der Zielwert (rötlicher Skalen-Strich) wird dabei meist mittig festgelegt. 

<br><p align="justify"><i>Lösung:</i> Um die maximale Grösse des Tachometers zu berechnen, wurde eine weitere Spalte in der SQL-View berechnet. Diese berechnet für jedes Issue jeweils den Max Wert aus entweder den geschätzten Stunden oder den effektiv aufgewendeten Stunden. Dieser wert wird mit einem Koeffizienten von 1.2 multipliziert und dem Visual als Referenzwert angegeben.

```python
# Code-Snippet 620
# Sprache: Microsoft SQL
# MaxTimeEstimatedOrLogged
# -----------------------------------------------------------------------------------------------------

CASE WHEN 
  COALESCE (dbo.gemini_issues.estimatehours, 0.0) * 1.0 + COALESCE (dbo.gemini_issues.estimateminutes, 0.0) / 60.0 > (COALESCE (dbo.gemini_issues.loggedhours, 0.0) * 1.0 + COALESCE (dbo.gemini_issues.loggedminutes, 0.0) / 60.0 ) 
THEN 
  ( (COALESCE (dbo.gemini_issues.estimatehours, 0.0) * 1.0 + COALESCE( dbo.gemini_issues.estimateminutes, 0.0) ) * 1.2 )
ELSE 
 ( ( COALESCE (dbo.gemini_issues.loggedhours, 0.0) * 1.0 + COALESCE (dbo.gemini_issues.loggedminutes, 0.0) / 60.0 ) * 1.2 )
END AS MaxTimeEstimateOrLogged
```

Die gesamte View kann nun mit folgendem Query erstellt werden:

```python
# Code-Snippet 700
# Sprache: Microsoft SQL
# SE_Power_BI Cockpit View
# -----------------------------------------------------------------------------------------------------

USE [SE_Gemini_PROD]
GO

/****** Object:  View [dbo].[SE_PowerBI]    Script Date: 5/28/2023 8:34:21 AM ******/
SET ANSI_NULLS ON
GO

SET QUOTED_IDENTIFIER ON
GO

CREATE OR ALTER VIEW [dbo].[SE_PowerBI]
AS
SELECT dbo.gemini_issues.fixedinversionid,
       dbo.gemini_versions.versionnumber,
       dbo.gemini_versions.versionname,
       dbo.gemini_projects.projectcode,
       dbo.gemini_projects.projectname,
       dbo.gemini_issuestatus.statusdesc,
       dbo.gemini_issues.summary,
       dbo.gemini_issues.issueid,
       dbo.gemini_issues.estimatehours,
       dbo.gemini_issues.estimateminutes,
       COALESCE (dbo.gemini_issues.estimatehours, 0.0) * 1.0 +
       COALESCE (dbo.gemini_issues.estimateminutes, 0.0) / 60.0               AS
       calcEstimated,
       dbo.gemini_issues.loggedhours,
       dbo.gemini_issues.loggedminutes,
       COALESCE (dbo.gemini_issues.loggedhours, 0.0) * 1.0 +
       COALESCE (dbo.gemini_issues.loggedminutes, 0.0) / 60.0                 AS
       calcLogged,
       ( ( COALESCE (dbo.gemini_issues.loggedhours, 0.0) * 1.0 +
             COALESCE (dbo.gemini_issues.loggedminutes, 0.0) / 60.0 ) -
         COALESCE (dbo.gemini_issues.estimatehours, 0.0) * 1.0 ) +
       COALESCE (dbo.gemini_issues.estimateminutes, 0.0) / 60.0               AS
       calcDiffHours,
       dbo.gemini_users.username,
       dbo.gemini_users.firstname,
       dbo.gemini_users.surname,
       dbo.gemini_users.emailaddress,
       dbo.gemini_users.roles,
       Components.components,
       dbo.gemini_issues.created,
       dbo.gemini_issues.closeddate,
       Datediff(day, dbo.gemini_issues.created, dbo.gemini_issues.closeddate) AS
       Durchlaufzeit,
       dbo.gemini_versions.versionid,
       dbo.gemini_users.active,
       CASE
         WHEN COALESCE (dbo.gemini_issues.estimatehours, 0.0) * 1.0 +
                     COALESCE (dbo.gemini_issues.estimateminutes, 0.0) / 60.0 >
              (
              COALESCE (dbo.gemini_issues.loggedhours, 0.0) * 1.0 +
                     COALESCE (
                     dbo.gemini_issues.loggedminutes, 0.0) / 60.0 ) THEN ( (
         COALESCE (dbo.gemini_issues.estimatehours, 0.0) * 1.0 +
COALESCE
(
                                                                      dbo.gemini_issues.estimateminutes, 0.0) ) * 1.2 )
  ELSE ( ( COALESCE (dbo.gemini_issues.loggedhours, 0.0) * 1.0 +
                    COALESCE (dbo.gemini_issues.loggedminutes, 0.0) / 60.0 ) *
         1.2 )
END                                                                    AS
       MaxTimeEstimateOrLogged
FROM   dbo.gemini_issues
       INNER JOIN dbo.gemini_issuestatus
               ON dbo.gemini_issues.issuestatusid =
                  dbo.gemini_issuestatus.statusid
       INNER JOIN dbo.gemini_projects
               ON dbo.gemini_issues.projectid = dbo.gemini_projects.projectid
       LEFT OUTER JOIN (SELECT String_agg(dbo.gemini_components.componentname,
                               ', ') AS
                               Components,
                               dbo.gemini_issuecomponents.issueid
                        FROM   dbo.gemini_components
                               INNER JOIN dbo.gemini_issuecomponents
                                       ON dbo.gemini_issuecomponents.componentid
                                          =
                                          dbo.gemini_components.componentid
                        GROUP  BY dbo.gemini_issuecomponents.issueid) AS
                       Components
                    ON Components.issueid = dbo.gemini_issues.issueid
       LEFT OUTER JOIN dbo.gemini_versions
                    ON dbo.gemini_issues.fixedinversionid =
                       dbo.gemini_versions.versionid
       LEFT OUTER JOIN dbo.gemini_issueresources
                    ON dbo.gemini_issues.issueid =
                       dbo.gemini_issueresources.issueid
       LEFT OUTER JOIN dbo.gemini_users
                    ON dbo.gemini_issueresources.userid =
                       dbo.gemini_users.userid 
GO
```
